```python
#Series 
import pandas as pd 
data =[1,2,3,4,5]
series = pd.Series(data)
print("Series\n",series)
print(type(series))
```

```python
##Series from Dictonary
data = {'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
#DataFrame: It will be having multiple rows and colums . Ex:database or excel
import pandas as pd
data={
    'Name':['Jon','Jack','Berlin'],
    'Age':[25,30,45],
    'City':['Bangalore','NewYork','Florida']
}
df=pd.DataFrame(data)
print(type(df))
```
